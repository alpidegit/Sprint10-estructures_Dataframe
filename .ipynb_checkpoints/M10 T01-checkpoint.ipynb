### a.1) http://quotes.toscrape.com con Beautiful Soup

1) Indentificamos los objetos que continen la información que queremos obtener, en la página web (Autor, el quote y sus Tags :

<small class="author" itemprop="author">Albert Einstein</small>
<a href="/author/Albert-Einstein">(about)</a>

**1)** Descargamos el contendio en html de la web a nuestro entorno Python:

```python
import requests
```

```python
# Este código realiza una solicitud HTTP GET a la URL proporcionada.
# Recupera los datos HTML que el servidor devuelve y almacena esos datos en un objeto de Python.
URL = "http://quotes.toscrape.com"
page = requests.get(URL).text  # para que me descargue todo el html en formato texto
print(page)
```

**2)** Creamos la "sopa", es decir, el arbol de objetos Python que representa al documento HTML, con Beautiful Soup:

```python
from bs4 import BeautifulSoup

# el parametro page.content se pasa a bs para analizar la pagina.
# el analizador html.parser se utiliza como analizador html para bs.
soup = BeautifulSoup(page, "html.parser") # si fuera xml, haríamos pi instal lxml, y aqui sería "lxml"
```

**3)** Buscar los elementos en la "sopa" (el arbol de objetos) y obtener la información deseada:

Empezaré solo buscando los autores:

```python
# busco todos los elementos "small" que tienen el atributo clase: "author"
autores = soup.find_all("small", class_ = "author") # pongo class_ porque es una palabra reservada y no se peude usar directamente como un parametro. POngo el guion bajo class_que es una convencion de BeatufulSoup para referirse a la clase CSS.
# Recorro los elementos de "autores" y obtengo el texto del autor
for author in autores:
    author_name = author.text
    print(author_name)
```

**Voy a buscar la frase y el autor y los tags:**

```python
quotes = soup.find_all("div", class_ = "quote")
for quote in quotes:
    cita = quote.find("span", class_="text")
    Autor = quote.find("small", class_="author")
    Tags = quote.find("div", class_="tags")
    print(cita.text.strip())   # añado .text para que solo coja textos de el elemento Bs
    print(Autor.text.strip())  # añado strip)= apra  eliminar los espacios antes y despues del texto
    print(Tags.text.strip())
    print()                    #Imprime una linea en blanco para separarlo del siguiente
```

### a.2) http://quotes.toscrape.com con Selenium

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
# Configuro el controlador web (en este caso, Chrome)
driver = webdriver.Chrome()

# Abro la página web
driver.get("http://quotes.toscrape.com")

# Espero unos segundos a que la página se cargue completamente
time.sleep(3)

# Encuentro todas las citas, autores y etiquetas en la página
citas = driver.find_elements(By.CLASS_NAME, "text")
autores = driver.find_elements(By.CLASS_NAME, "author")
etiquetas = driver.find_elements(By.CLASS_NAME, "tags")

# Imprimo los datos encontrados
for i in range(len(citas)):
    print(citas[i].text.strip())
    print(autores[i].text.strip())
    print(etiquetas[i].text.strip())
    print()
    
# Cierro la página web y termino la sesión del navegador
driver.quit()
```

### b.1) www.wikipedia.es (fes alguna cerca primer i escrapeja algun contingut) con BeautifulSoup

Voy a hacer web scraping en la web de Wikipedia que define "web scraping". Sacaré el listado de herramientas que comenta con su url:

Web: https://es.wikipedia.org/wiki/Web_scraping

```python
URL = "https://es.wikipedia.org/wiki/Web_scraping"
page = requests.get(URL).text  #me devuelve el contenido html en formato texto
soup = BeautifulSoup(page, "lxml")  # USo lxml por cambiar sabiendo que ademas es mas potente que html.parser pues trabaja ocn html y xml
```

```python
#Capturo la parte del codigo que me interesa
datos = soup.find("div", class_ = "columns")
# Inicializo listas para almacenar los títulos y los enlaces
titulos = []
enlaces = []

# Encuentro los elementos <a> dentro datos
elementos_a = datos.find_all("a")

# Recorremos el elemento <a> para obtener titulos y enlaces web
for elemento in elementos_a:
    titulo = elemento["title"]  # Obtengo el texto del enlace (el título)
    enlace_url = elemento["href"]  # Obtenego el valor del atributo 'href' (el enlace URL)
    titulos.append(titulo)
    enlaces.append(enlace_url)

# Imprimo los títulos y enlaces
for i in range(len(titulos)):
    print(f"Título: {titulos[i]}")
    print(f"Enlace: {enlaces[i]}")
    print()
```

### b.2) www.wikipedia.es (fes alguna cerca primer i escrapeja algun contingut) con Selenium

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome() # Configuro el controlador web (en este caso, Chrome)
driver.get("https://www.wikipedia.org/")
time.sleep(2)

#voy a indentificar el campo de busqueda en la pag web
#indetifco el campo de busqueda y el boton de enviar
campo_busqueda = driver.find_element(By.ID, "searchInput") 
boton_busqueda = driver.find_element(By.CSS_SELECTOR, 'button[class="pure-button pure-button-primary-progressive"][type="submit"]')  

# voy a enviar una busqueda
campo_busqueda.send_keys("Web scraping")
boton_busqueda.click()
time.sleep(2)

# Ahora que ya estoy en la web que he buscado, extraigo los datos que me interesan
        #(los que he buscado en el ejercicio b.1)
datos1 = driver.find_element(By.CSS_SELECTOR, "div.columns") 

# Inicializo listas para almacenar los títulos y los enlaces
titulos = []
enlaces = []

# Encuentro los elementos <a> dentro datos
elementos_a = datos1.find_elements(By.TAG_NAME, "a")

# Recorremos el elemento <a> para obtener titulos y enlaces web
for elemento in elementos_a:
    titulo = elemento.get_attribute("title")  # Obtengo el texto del enlace (el título)
    enlace_url = elemento.get_attribute("href")  # Obtenego el valor del atributo 'href' (el enlace URL)
    titulos.append(titulo)
    enlaces.append(enlace_url)

# Imprimo los títulos y enlaces
for i in range(len(titulos)):
    print(f"Título: {titulos[i]}")
    print(f"Enlace: {enlaces[i]}")
    print()
    
driver.quit()
```

### c.1) https://www.bolsamadrid.es con Selenium

Voy a identificar y clickar el Boton IBEX 35:

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Abro la Web general ("https://www.bolsamadrid.es")
driver = webdriver.Chrome() # Configuro el controlador web (en este caso, Chrome)
driver.get("https://www.bolsamadrid.es")
time.sleep(2)

# Creo una instancia de WebDriverWait con un tiempo máximo de espera de 10 segundos
wait = WebDriverWait(driver, 10)

# Quito la ventana emergente de cookies si está presente
try:
    cookies_popup = wait.until(EC.presence_of_element_located((By.ID, "onetrust-banner-sdk")))
    cookies_popup.find_element(By.XPATH, "//button[contains(text(), 'Aceptar')]").click()
except:
    pass

# indetifico el boton IBEX 35 (busco en el elemento <a> con la classe "section-buttons-button" que contiene "ibex-35" )
                            # Atencion: si pongo "" tengo que poner fuera '', o al reves
boton_IBEX35 = driver.find_element(By.CSS_SELECTOR, 'a.section-buttons-button[href*="ibex-35"]')

#Clicko en el boton del IBEX 35
boton_IBEX35.click()
time.sleep(5)
```

Como me abre la pagina sobre la anterior y no en una nueva ventana, el elemento dirver automáticamente hace referencia a la nueva pagina.  
Si me lo abriera en una ventana adicional tendría que cambiar el contexto entre ventanas con el codigo siguiente:  

Obtendría las identificaciones de todas las ventanas abiertas  
**ventana_principal = driver.current_window_handle  
ventanas_abiertas = driver.window_handles**  

Cambiaría al contexto de la nueva ventana  
**nueva_ventana = [ventana for ventana in ventanas_abiertas if ventana != ventana_principal][0]  
driver.switch_to.window(nueva_ventana)**   
 y si quisiera volver al contexto de la ventana ppal:   **driver.switch_to.window(ventana_principal)**  

Ahora estaría en el contexto de la nueva ventana y puedes realizar el web scraping en ella

**Obtengo en la web del IBEX 35 la Fecha y hora de extraccción:**

```python
# Espero a que el elemento que busco este rpesente en la cabecera
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'head[data-hora-act]')))

# Obtengo el elemento de cabecera
head_element = driver.find_element(By.CSS_SELECTOR, 'head[data-hora-act]')

# Estraigo el valor de atributo "data-hora-act"
fechayhora_actualizacion = head_element.get_attribute("data-hora-act")

# Imprimo el valor
print("Fecha y hora de actualización:  ", fechayhora_actualizacion)
```

**Obtengo los titulos de la tabla de valores:**

```python
# Primero encuentro la cabecera <thead> con sus elementos (esta corresponde a la cabecera de al tabla)
thead_element = driver.find_element(By.CSS_SELECTOR, "thead")

# Encuentro todos los elementos <th> dentro de el elemento <thead>
th_elements = thead_element.find_elements(By.TAG_NAME, "th")

# Inicializo una lista vacia de titulos e itero sobre los elementos <th> sacando el texto de cada titulo
Titulos = []

#  Itero sobre la lista de elementos <th> y extraigo el texto que corresponde a los titulos
for th_element in th_elements:
    titulo = th_element.text.strip()
    Titulos.append(titulo)

print(Titulos)
```

Obtengo los valores con los resultados para cada columna segun los titulos anteriores:

```python
# Primero encuentro la tabla cuya clase es "shares-table"
tabla = driver.find_element(By.CLASS_NAME, "shares-table")

# obtengo todas las filas <tr> de la tabla
filas = tabla.find_elements(By.TAG_NAME, "tr")

# Hago un diccionario llamado Valores con las claves de la lista Titulos para almacenar los datos de la tabla.
# Cada clave (titulo) se inicializa con una lista vacía para almacenar los valores correspondientes de esa columna.
Valores = {titulo: [] for titulo in Titulos}

# Extraigo los datos de cada fila de la tabla y los agrego a la lista "Valores"
for fila in filas[1:]:  # Ignoro la primera fila de los titulos de columnas
    celdas = fila.find_elements(By.TAG_NAME, 'td')  # para cada fila obtengo las celdas <td>
#se recorren las celdas y sus valores se agregan a la lista correspondiente en el diccionario Valores, utilizando el nombre de la columna.
    for i, titulo in enumerate(Titulos): # Utilizo enumerate para obtener el indice y el valor de los Titulos
        Valores[titulo].append(celdas[i].text.strip())

# Cierro la ventana
driver.quit()

# Creo un DataFrame con la tabla
df = pd.DataFrame(Valores)
df
```

Información extraída de: https://www.bolsamadrid.es

# Contenido

35 filas y 9 columnas (siendo la primera una numeracion correlativa sin mass significado). 

**Descripción de las colummas:**

- Índice: Es un identificador único o código que representa a una acción o valor específico en el mercado de valores. Cada acción o valor cotizado tiene su propio índice para identificarlo de manera única.

- Último: Representa el último precio al cual se negoció la acción o valor en el mercado. Es el precio más reciente registrado para la acción.

- % Dif.: Es el porcentaje de diferencia entre el precio actual de la acción y el precio de cierre del día anterior. Indica la variación porcentual del precio desde el último cierre.

- Máximo: Es el precio más alto alcanzado por la acción o valor durante el día de negociación actual.

- Mínimo: Es el precio más bajo alcanzado por la acción o valor durante el día de negociación actual.

- Fecha: Es la fecha en la que se están registrando los datos de cotización de la acción.

- Hora: Es la hora a la que corresponde el precio registrado para la acción.

- % Dif. Año: Representa el porcentaje de diferencia entre el precio actual de la acción y el precio de cierre al final del año anterior. Indica la variación porcentual del precio desde el final del año pasado.

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Abro la Web general de ("https://www.casadellibro.com/")
driver = webdriver.Chrome()  # Configuro el controlador web (en este caso, Chrome)
driver.get("https://www.casadellibro.com/")

# Espero a que cargue la web y el banner de cookies
wait = WebDriverWait(driver, 5)

# Quito la ventana emergente de cookies si está presente
try:
    cookies_popup = wait.until(EC.presence_of_element_located((By.ID, "onetrust-button-group-parent")))
    cookies_popup.find_element(By.ID, "onetrust-accept-btn-handler").click()
except:
    pass

# Identifico el botón de "Lo más leído" y haz clic en él
elemento_lo_mas_leido = driver.find_element(By.CSS_SELECTOR, 'a[title="Ver más"]')
elemento_lo_mas_leido.click()

# Espero a que cargue la web nuevamente
wait.until(EC.presence_of_element_located((By.CLASS_NAME, "compact-product")))

# Listado de géneros con sus ID
Generos = ["masleidosficcion23", "masleidosnoficcion23", "comic23", "juvenil23", "infantil23"]

# Listas para almacenar la información de los libros
genero = []
titulos = []
formatos = []
autores = []
precios = []

# Itero a través de los géneros y obtiene los libros para cada uno
for gen in Generos:
    # Encuentro la sección de los libros del género más leídos de la semana
    seccion_element = driver.find_element(By.ID, gen)

    # Encuentro todos los contenedores de libros del genero dentro de la sección
    contenedores_libros = seccion_element.find_elements(By.CSS_SELECTOR, "div.compact-product")

    # Itero a través de los contenedores de libros y extraigo la información
    for contenedor in contenedores_libros:
        genero.append(gen)

        # Extrigo el título del libro
        titulo_element = contenedor.find_element(By.CLASS_NAME, "compact-product-title")
        titulos.append(titulo_element.text.strip())

        # Extraigo el formato del libro
        formato_element = contenedor.find_element(By.CLASS_NAME, "feature")
        formatos.append(formato_element.text.strip())

        # Extraigo el autor del libro
        autor_element = contenedor.find_element(By.CLASS_NAME, "compact-product-authors")
        autores.append(autor_element.text.strip())

        # Extraigo el precio del libro
        precio_element = contenedor.find_element(By.CLASS_NAME, "main-price")
        precios.append(precio_element.text.strip())

# Crea un DataFrame con la información de los libros
df_todos_libros = pd.DataFrame({
    "Género": genero,
    "Título": titulos,
    "Formato": formatos,
    "Autor": autores,
    "Precio": precios
})

# Cerrar el navegador
driver.quit()
```

## 3.2) Casa del Libro (https://www.casadellibro.com/) con Scrapy 

Tras instalar el scrapy, creo una carpeta en un entorno virtual.    
En Anaconda, en el directorio envs he creado el nuevo directorio "cassadellibro" y el spider **casadellibro.py**.
Finalmente lo he hecho correr con **scrapy crawl casadellibro** en una ventana de comando.

Este es el codigo del spider:

```python
import scrapy
import pandas as pd

class CasadellibroSpider(scrapy.Spider):
    name = "casadellibro"
    allowed_domains = ["casadellibro.com"]
    start_urls = ["https://www.casadellibro.com/"]

    def parse(self, response):
        # Your existing code for handling cookies_popup goes here...

        # Identifico el botón de "Lo más leído" y hago clic en él
        elemento_lo_mas_leido = response.css('a[title="Ver más"]')
        if elemento_lo_mas_leido:
            yield response.follow(elemento_lo_mas_leido, callback=self.parse_mas_leidos)

    def parse_mas_leidos(self, response):
        # Creo las listas para almacenar la información de genero de los libros
        generos = ["masleidosficcion23", "masleidosnoficcion23", "comic23", "juvenil23", "infantil23"]

        # Lista para almacenar todos los datos de los libros
        Libros = []

        for gen in generos:
            # Encuentro la sección de los libros del género más leídos de la semana
            seccion_element = response.css(f'#{gen}')

            # Encuentro todos los contenedores de libros del genero dentro de la sección
            contenedores_libros = seccion_element.css("div.compact-product")

            # Itero a través de los contenedores de libros y extraigo la información
            for contenedor in contenedores_libros:
                titulo = contenedor.css(".compact-product-title::text").get().strip() 
                formato = contenedor.css(".feature::text").get().strip()
                autor = contenedor.css(".compact-product-authors::text").get().strip()
                precio = contenedor.css(".main-price::text").get().strip()

                 # :: es un selector que permite seleccionar algo de lo buscado, en este caso, el texto deelemento buscado con CSS
              
            # Agrego los datos a la lista de todos los libros
                Libros.append({
                    "Género": gen,
                    "Título": titulo,
                    "Formato": formato,
                    "Autor": autor,
                    "Precio": precio
                })

        # Creo el DataFrame con toda la información de los libros
        df_libros = pd.DataFrame(Libros)

        # Guardo el DataFrame en un archivo CSV
        df_libros.to_csv("libros_casadellibro.csv", index=False)
```
